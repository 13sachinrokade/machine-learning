```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

d0=pd.read_csv("train.csv")
l=d0["label"]
d=d0.drop("label",axis=1)
labels=l.head(15000) #some sample of first 15k
data=d.head(15000)
print(data.shape)
#data preprocessing
from sklearn.preprocessing import StandardScaler
standardized_data=StandardScaler().fit_transform(data)
print(standardized_data.shape)
```

```python
from sklearn.manifold import TSNE
data_1000=standardized_data[0:1000,:] #taking only 1k because 15k takes nearly 20min
labels_1000=labels[0:1000]
model=TSNE(n_components=2,random_state=0)
#default perplexity=30#default leaning rate=200
#default  stepsize or noof iterations=1000
tsne_data=model.fit_transform(data_1000)
tsne_data1=np.vstack((tsne_data.T,labels_1000)).T
tsne_df=pd.DataFrame(data=tsne_data1,columns=("dim1","dim2","label"))
sns.FacetGrid(tsne_df,hue="label",size=6).map(plt.scatter,"dim1","dim2").add_legend()
plt.show()
```

```python
model=TSNE(n_components=2,random_state=0,perplexity=50)
#default perplexity=30#default leaning rate=200
#default  stepsize or noof iterations=1000
tsne_data=model.fit_transform(data_1000)
tsne_data1=np.vstack((tsne_data.T,labels_1000)).T
tsne_df=pd.DataFrame(data=tsne_data1,columns=("dim1","dim2","label"))
sns.FacetGrid(tsne_df,hue="label",size=6).map(plt.scatter,"dim1","dim2").add_legend()
plt.title("perplexity=50")
plt.show()
```

```python
model=TSNE(n_components=2,random_state=0,perplexity=50,n_iter=5000)
#default perplexity=30#default leaning rate=200
#default  stepsize or noof iterations=1000
tsne_data=model.fit_transform(data_1000)
tsne_data1=np.vstack((tsne_data.T,labels_1000)).T
tsne_df=pd.DataFrame(data=tsne_data1,columns=("dim1","dim2","label"))
sns.FacetGrid(tsne_df,hue="label",size=6).map(plt.scatter,"dim1","dim2").add_legend()
plt.title("perplexity=50 ans iterations=5000")
plt.show()
```
