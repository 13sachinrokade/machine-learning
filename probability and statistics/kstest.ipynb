```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

x=stats.norm.rvs(size=1000)
sns.set_style("whitegrid")
sns.kdeplot(np.array(x),bw=0.5)
plt.show()
```

```python
np.mean(x)
```

```python
np.var(x)
```

```python
stats.kstest(x,"norm")
```

```python
stats.kstest(x,"uniform")
```
