```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
std_normal=np.random.normal(loc=0,scale=1,size=1000)
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
measure=np.random.normal(loc=20,scale=5,size=10)
stats.probplot(measure,dist="norm",plot=pylab)
pylab.show()
```

for less no of samples we cant estimate

```python
measure=np.random.normal(loc=20,scale=5,size=1000)
stats.probplot(measure,dist="norm",plot=pylab)
pylab.show()
```

for more sample if all points lie on the line then disturibution on x and y axis are same


```python
measure=np.random.normal(loc=20,scale=5,size=1000)
stats.probplot(measure,dist="uniform",plot=pylab)
pylab.show()
```

uniform vs normal many points are not lying on the line so they are not same

```python
measure=np.random.uniform(low=-1,high=1,size=1000)
stats.probplot(measure,dist="norm",plot=pylab)
pylab.show()
```

```python
measure=np.random.uniform(low=1,high=5,size=1000)
stats.probplot(measure,dist="uniform",plot=pylab)
pylab.show()
```
